Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Import the necessary packages
import pandas as pd
import numpy as np

# Read in the data as a pandas df
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()

```

```python
# Which NYC schools have the best math results?

# Define best math score
best_score = schools["average_math"] >= 640

# Filter the dataset by scores and subset needed columns
best = schools[best_score][["school_name", "average_math"]]

# Sort the dataset for top scores by math score 
best_math_schools = best.sort_values("average_math", ascending=False)
print(best_math_schools.head())
```

```python
# What are the top 10 performing schools based on the combined SAT scores?

# Create total_SAT columns for combine SAT scores
schools["total_SAT"] = schools["average_reading"] + schools["average_writing"] + schools["average_math"]
#schools["total_SAT"] = schools[["average_reading", "average_writing", "average_math"]].sum(axis=1)

# Sort the column
schools_sorted = schools.sort_values("total_SAT", ascending=False)

# Grab the top scores
top_10_schools = schools_sorted.iloc[:10, [0,7]]
print(top_10_schools)
```

```python
# Which single borough has the largest standard deviation in the combined SAT score?

# Preview data
schools.head()

# Create a grouped df (borough and total_SAT) with mean & sd aggregation
boroughs_me_sum = schools.groupby("borough")["total_SAT"].agg([np.mean, np.std]).round(2)
print(boroughs_me_sum)
```

```python
# conti..
# Insert new column to show number of schools in each grouped borough
boroughs_me_sum["num_schols"] = schools["borough"].value_counts()

# Rename the index column and reset it back into the df
boroughs_rename = boroughs_me_sum.rename_axis("borough").reset_index()

# Rename the column with appropriate names
boroughs_rename.rename(columns = {"mean": "average_SAT", "std": "std_SAT", "num_schols": "num_schools"}, inplace=True)

# Fiter the df by the max std
largest_std_dev = boroughs_rename[boroughs_rename["std_SAT"] == boroughs_rename["std_SAT"].max()]
print(largest_std_dev)

```
